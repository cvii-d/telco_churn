```python
import pandas as pd

#df = pd.read_csv('Telco-Customer-Churn.csv')
df = pd.read_csv('https://raw.githubusercontent.com/cvii-d/telco_churn/refs/heads/main/Telco-Customer-Churn.csv')
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (replace with actual dataset)

# Plot churn rates for core services
service_columns = ['PhoneService', 'MultipleLines', 'InternetService']
for column in service_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=column, hue='Churn', palette='Set2')
    plt.title(f'Churn Rates by {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(title='Churn', loc='upper right')
    plt.show()
```

```python
additional_services = ['OnlineSecurity', 'DeviceProtection', 'TechSupport']

# Plot churn rates for additional services
for column in additional_services:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=column, hue='Churn', palette='Set2')
    plt.title(f'Churn Rates by {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(title='Churn', loc='upper right')
    plt.show()
```

```python
streaming_services = ['StreamingTV', 'StreamingMovies']


# Plot churn rates for streaming services
for column in streaming_services:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=column, hue='Churn', palette='Set2')
    plt.title(f'Churn Rates by {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(title='Churn', loc='upper right')
    plt.show()
```

```python
# Grouped bar chart for churn rates by contract type
contract_churn = df.groupby(['Contract', 'Churn']).size().unstack()
contract_churn.plot(kind='bar', stacked=True, figsize=(8, 6), color=['green', 'red'])
plt.title('Churn Rates by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.legend(title='Churn', loc='upper right')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Convert 'TotalCharges' to numeric (if it contains non-numeric values)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Encode categorical columns as numeric
binary_columns = ['Churn', 'Partner', 'Dependents', 'PhoneService', 
                  'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                  'TechSupport', 'StreamingTV', 'StreamingMovies', 
                  'PaperlessBilling']

df['gender'] = df['gender'].apply(lambda x: 1 if x == 'Male' else 0)
df['InternetService'] = df['InternetService'].map({'DSL': 1, 'Fiber optic': 2, 'No': 0})

for col in binary_columns:
    df[col] = df[col].apply(lambda x: 1 if x == 'Yes' else 0)

# Select all numeric columns automatically
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Plot the heatmap
plt.figure(figsize=(25, 15))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Numeric Features')
plt.savefig('my_plot.png')
plt.show()
```
